```python
import pandas as pd
#read the data1

data1=pd.read_csv('GSE262814_raw_counts_GRCh38.p13_NCBI.tsv', sep='\t')

#load the data1
data1
```

```python
data2=pd.read_csv('GSE262815_raw_counts_GRCh38.p13_NCBI.tsv', sep='\t')

#load the data2
data2
```

```python
data3=pd.read_csv('GSE262828_raw_counts_GRCh38.p13_NCBI.tsv', sep='\t')

#load the data3
data3
```

```python
merged_data=pd.merge(data1,data2,on='GeneID')
final_data=pd.merge(merged_data,data3,on='GeneID')

final_data
```

```python
# Remove leading and trailing whitespace from column names in the DataFrame
final_data.columns = final_data.columns.str.strip()

final_data
```
